```python
import yfinance as yf
import pandas as pd
import numpy as np
from datetime import datetime
import plotly.express as px
```

```python
def add_year(date, years):
    date = datetime.strptime(date, '%Y-%m-%d')
    return date.replace(year = date.year + years).strftime("%Y-%m-%d")

def add_month(date, months):
    date = datetime.strptime(date, '%Y-%m-%d')
    if date.month + months > 12:
       return date.replace(month=date.month + months - 12, year =  date.year+1).strftime("%Y-%m-%d")
    else:
        return date.replace(month = date.month + months).strftime("%Y-%m-%d")
```

```python
date_1 = '2019-01-01'
date_2 = '2021-02-01'
mrcdo = 'BTC-USD'
# df = yf.Ticker(mrcdo).history(start=date_1, end=add_year(date_1, 1))[['Close']].rename(columns={'Close':mrcdo})
# prc_0 = df.loc[date_1][0]
# df
rpv_1 = 20
rpv_2 = 70

rpvs = [x / 100 for x in range(rpv_1,rpv_2+1)]
zero = [0 for x in range(rpv_1, rpv_2+1)]
rpvs = dict(zip(rpvs, zero))
```

```python
cont = 0
while datetime.strptime(date_1, '%Y-%m-%d') <= datetime.strptime(date_2, '%Y-%m-%d'):
    df = yf.Ticker(mrcdo).history(start=date_1, end=add_year(date_1, 1))[['Close']].rename(columns={'Close':mrcdo})
    prc_0 = df.loc[date_1][0]
    for i in rpvs:
        loan= i * prc_0
        df['LTV'] = loan/df[mrcdo]
        if df[df['LTV']>=.8].empty:
            pass
        else:
            rpvs[i] += 1

    date_1 = add_month(date_1, 1)
    cont +=1
```
